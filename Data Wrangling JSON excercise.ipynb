```python
data = json.load((open('data/world_bank_projects.json'))) #load input data
themes_df = json_normalize(data, 'mjtheme_namecode',['_id','countryname'])  # normalize dataframe to split multiple codes/names per project id.
csort_df = themes_df.set_index(['code']).sort_index()  #sort df by column "code"
csort_df['name'] = csort_df.name.replace('',method='ffill')  #forward fill name column to fill blank spaces 
csort_df=csort_df.reset_index()      #reset df to make "code" a column
csort_df['_id'] = csort_df['_id'].astype(str)  #set type from dictionary to string
csort_df=csort_df.drop_duplicates(['_id','code'], keep='first')  #remove duplicate entry for same project id (but it's not working.)

q2_df = csort_df.groupby(['code','name']).size().sort_values(ascending=False).head(10)#group df by size of code and sort by descending

q2_df
```
